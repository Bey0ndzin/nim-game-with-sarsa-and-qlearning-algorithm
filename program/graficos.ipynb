```python
from nim import SARSA, QLearning, train, play_against_eachother
import matplotlib.pyplot as plt
import random
```

```python
ai = SARSA()
sarsa = train(player = ai, n_episodes = 1000)

ai = QLearning()
qlearning = train(player = ai, n_episodes = 1000)

n_jogos = 100

qtdVictorySarsa = 0
victorySarsa = []
qtdVictoryQLearning = 0
victoryQLearning = []

for jogo in range(n_jogos):
    vencedor = play_against_eachother(sarsa, qlearning)
    if vencedor == "SARSA":
        qtdVictorySarsa += 1
    else:
        qtdVictoryQLearning += 1

    victorySarsa.append(qtdVictorySarsa)
    victoryQLearning.append(qtdVictoryQLearning)

# Criando o gráfico de linhas
plt.plot(range(1, n_jogos + 1), victorySarsa, label="SARSA", color="blue", linewidth=2)
plt.plot(range(1, n_jogos + 1), victoryQLearning, label="QLearning", color="green", linewidth=2)

# Adicionando título e rótulos aos eixos
plt.title('Vitórias Acumuladas ao Longo dos Jogos')
plt.xlabel('Número de Jogos')
plt.ylabel('Número de Vitórias')
plt.legend()

# Exibindo o gráfico
plt.show()
```
